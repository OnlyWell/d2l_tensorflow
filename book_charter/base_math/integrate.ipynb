```python
'''
    积分运算
'''
import numpy as np
import scipy.integrate as integrate
import numpy as np
import math

f = lambda x : x # 原函数为 y = (1/2) * x^2
i,e = integrate.quad(f,0,1)

print('积分值:' + str(i))
print('误差:' + str(e))
```

```python
#画图表示
import matplotlib.pyplot as plt

#样本点
x = range(0,11)

#作图
plt.plot(x,f(x),color='purple')

#面积
area = np.arange(0,3,1/20)
plt.fill_between(area,f(area),color='green')

#设定图形属性
plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid()
plt.show()
```

```python
#正态分布
import scipy.integrate as integrate
import numpy as np
import math

#平均数
mean = 0
std = 1 # 标准差

#正态分布的概率密度函数
f = lambda x : (1 / (std * math.sqrt(2 * math.pi))) * math.exp(-(x - mean)**2 / (2 * std**2))

#积分,从负无穷大到正无穷大
i,e = integrate.quad(f,-np.inf,np.inf)

print('累计概率:', round(i,2))
print('误差:', str(e))
```

```python
# ±1的置信区间
i,e = integrate.quad(f,-1,1)
print('置信区间:', round(i,3))
print('误差:', str(e))

#±1.96的置信区间
i,e = integrate.quad(f,-1.96,1.96)
print('置信区间:', round(i,3))
print('误差:', str(e))

# ±2的置信区间
i,e = integrate.quad(f,-2,2)
print('置信区间:', round(i,3))
print('误差:', str(e))

# ±3的置信区间
i,e = integrate.quad(f,-3,3)
print('置信区间:', round(i,3))
print('误差:', str(e))

```

```python
'''
    绘制随机数及直方图
'''
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 使用random产生标准正态分布的10000个随机数
x = np.random.randn(10000)

# 绘制平滑曲线
sns.kdeplot(x)

# 设定图形属性
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# 对±1、±2、±3倍标准差画虚线
for std_dev in [1, 2, 3]:
    plt.axvline(x=-std_dev, color='red', linestyle=':')
    plt.axvline(x=std_dev, color='red', linestyle=':')

# 标注标准差概率
plt.annotate(text='', xy=(-1, 0.25), xytext=(1, 0.25), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='68.3%', xy=(0, 0.26), xytext=(0, 0.26))

plt.annotate(text='', xy=(-2, 0.05), xytext=(2, 0.05), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='95.4%', xy=(0, 0.06), xytext=(0, 0.06))

plt.annotate(text='', xy=(-3, 0.01), xytext=(3, 0.01), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='99.7%', xy=(0, 0.02), xytext=(0, 0.02))

plt.show()
```
