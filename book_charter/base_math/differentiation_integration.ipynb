```python
'''
    f(x) = -10*x**2 + 100x +5,求最大值
'''
import matplotlib.pyplot as plt
import numpy as np

#定义函数
def f(x):
    return -10*x**2 + 100*x +5

#一阶导数
def df(x):
    return -20*x + 100

#设置样本点
x = list(range(-0, 11,1))
y = [f(i) for i in x]

#一阶导数的样本点
yd = [df(i) for i in x]

#画二次曲线
plt.plot(x,y,color='blue',linewidth=2)
#画一阶导数
plt.plot(x,yd,color='green',linewidth=2)

#画三个点的斜率(2,5,8)
x1 = 2
x2 = 5
x3 = 8

plt.plot([x1 - 1,x1 + 1],[f(x1) -(df(x1)),f(x1) + (df(x1))],color='red',linewidth=3)
plt.plot([x2 - 1,x2 + 1],[f(x2) -(df(x2)),f(x2) + (df(x2))],color='red',linewidth=3)
plt.plot([x3 - 1,x3 + 1],[f(x3) -(df(x3)),f(x3) + (df(x3))],color='red',linewidth=3)

#最大值
plt.axvline(5)

plt.grid()
plt.show()
```

```python
'''
    绘制f(x)=x**2 + 2*x +7,求最小值
'''

#定义函数
def f(x):
    return x**2 + 2*x +7
#一阶导数
def df(x):
    return 2*x + 2
#设置样本点
x = list(range(-12,11))
y = [f(i) for i in x]

#一阶导数的样本点
dy = [df(i) for i in x]

#绘制曲线
plt.plot(x,y,color='blue',linewidth=2)

#绘制一阶导数的曲线
plt.plot(x,dy,color='green',linewidth=2)

#取4个样本点
x1 = -5
x2 = -1
x4 = 5

#绘制四个点的斜率
plt.plot([x1 - 1,x1 + 1],[f(x1) -(df(x1)),f(x1) + (df(x1))],color='red',linewidth=3)
plt.plot([x2 - 1,x2 + 1],[f(x2) -(df(x2)),f(x2) + (df(x2))],color='red',linewidth=3)
plt.plot([x4 - 1,x4 + 1],[f(x4) -(df(x4)),f(x4) + (df(x4))],color='red',linewidth=3)

#取最小值(斜率为0)
plt.axvline(
    x=-1,
    color='blue',     # 线颜色
    linestyle=':',    # 线型（虚线）
    linewidth=2,      # 线宽
    alpha=0.77,        # 透明度（0-1）
    label='Critical Point'  # 图例标签
)
plt.legend()  # 显示图例

plt.grid()
plt.show()
```

```python
'''
    绘制三次曲线方程f(x)=x**3 -2*x+100,求最小值
'''
#定义函数
def f(x):
    return x**3 -2*x+100

#一阶导数
def df(x):
    return 3*x**2 -2
#二阶导数
def ddf(x):
    return 6*x

#设置样本点
x = list(range(-10,11))
y=[f(i) for i in x]

#设置一阶导数的样本点
dy=[df(i) for i in x]

#设置二阶导数的样本点
ddy=[ddf(i) for i in x]

#绘制函数的曲线
plt.plot(x,y,color='blue',linewidth=2)
#绘制一阶导数的曲线
plt.plot(x,dy,color='green',linewidth=2)
#绘制二阶导数的曲线
plt.plot(x,ddy,color='red',linewidth=2)

#取点
x1 = -5
x2 = 0
x3 = 4

#绘制三个点的一阶导数
plt.plot([x1 - 1,x1 + 1],[f(x1) -(df(x1)),f(x1) + (df(x1))],color='red',linewidth=3)
plt.plot([x2 - 1,x2 + 1],[f(x2) -(df(x2)),f(x2) + (df(x2))],color='red',linewidth=3)
plt.plot([x3 -1,x3 + 1],[f(x3) -(df(x3)),f(x3) + (df(x3))],color='red',linewidth=3)

#绘制三个点的二阶导数
plt.plot([x1 - 1,x1 + 1],[df(x1) -(ddf(x1)),df(x1) + (ddf(x1))],color='yellow',linewidth=3)
plt.plot([x2 - 1,x2 + 1],[df(x2) -(ddf(x2)),df(x2) + (ddf(x2))],color='yellow',linewidth=3)
plt.plot([x3 - 1,x3 + 1],[df(x3) -(ddf(x3)),df(x3) + (ddf(x3))],color='yellow',linewidth=3)

#绘制最小值
plt.axvline(
    x=0,
    color='blue',     # 线颜色
    linestyle=':',    # 线型（虚线）
    linewidth=2,      # 线宽
    alpha=0.77,        # 透明度（0-1）
    label='Critical Point'  # 图例标签
)

plt.legend()
plt.grid()
plt.show()

```

```python
'''
    Sympy库直接支持微积分函数的计算
'''
from sympy import *

x = Symbol('x')

#1.f(x)为常数
f = 0 *x + 5

yprime = diff(f,x)
print(yprime)

#2.C(f(x))
f = 2 * x + 5
yprime = diff(f,x)
print(yprime)

f = 3 * x **4 + 2 * x **3 + 5 * x **2 + 2 * x + 5
yprime = diff(f,x)
print(yprime)

print("===============\n")
#d[f(x)g(x)]/dx
f = x**2
g = x**3

y1 = f * g
yprime1 = y1.diff(x)
print(f'yprime1:{yprime1}')

#
yprime2 = f.diff(x) * g + f * g.diff(x)
print(f'yprime2:{yprime2}')

print(yprime1 == yprime2)
```

```python
'''
    微分的乘积规则和链式法则
    案例1:
'''
from sympy import *

x = Symbol('x')

#d[f(x)g(x)]/dx
f = x ** 2
g = x ** 3
y1 = f * g
yprime1 = y1.diff(x)

# f'(x)g(x) + f(x)g'(x)
yprime2 = f.diff(x) * g + f * g.diff(x)
print(yprime2)

# 比较导数
print(yprime1 == yprime2)
```
