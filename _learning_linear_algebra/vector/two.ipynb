```python
import math

#向量的加减法
import numpy as np
import matplotlib.pyplot as plt

v = np.array([2,1])
v1 = np.array([2,1]) + 2
v2 = np.array([2,1]) - 2

#原点
origin = [0],[0]

#画有箭头的线
plt.quiver(*origin, *v1,scale=10,color='r')
plt.quiver(*origin, *v,scale=10,color='b')
plt.quiver(*origin, *v2,scale=10,color='g')

plt.annotate('original vector', (0.025,0.01), xycoords = 'data', fontsize=16)

#作图
plt.axis('equal')
plt.grid()

plt.xticks(np.arange(-0.05,0.06,0.01),labels= np.arange(-5,6,1))
plt.yticks(np.arange(-3,5,1)/100,labels= np.arange(-3,5,1))
plt.show()
```

```python
#向量乘除法
v = np.array([2,1])
v1 = np.array([2,1]) * 2
v2 = np.array([2,1]) / 2

#绘制原点
origin = [0],[0]

#画有箭头的线
plt.quiver(*origin, *v1,scale=10,color='r')
plt.quiver(*origin, *v,scale=10,color='b')
plt.quiver(*origin, *v2,scale=10,color='g')

#图中加注释说明
plt.annotate('original vector', (0.025,0.008), xycoords = 'data', color = 'b', fontsize=16)

#作图
plt.axis('equal') #x,y的刻度相同
plt.grid()

plt.xticks(np.arange(-0.05,0.06,0.01),labels= np.arange(-5,6,1))
plt.yticks(np.arange(-3,5,1)/100,labels= np.arange(-3,5,1))
plt.show()
```

```python
#向量之间的加减
v = np.array([2,1])
s = np.array([-3,2])
v1 = v + s

#绘制原点
origin = [0],[0]

#画有箭头的线
plt.quiver(*origin, *v,scale=10,color='r')
plt.quiver(*origin, *s,scale=10,color='b')
plt.quiver(*origin, *v1,scale=10,color='g')

#图中加注释说明
plt.annotate('original vector', (0.025,0.008), xycoords = 'data', color = 'b', fontsize=16)

#作图
plt.axis('equal') #x,y的刻度相同
plt.grid()

plt.xticks(np.arange(-0.05,0.06,0.01),labels= np.arange(-5,6,1))
plt.yticks(np.arange(-3,5,1)/100,labels= np.arange(-3,5,1))
plt.show()
```

```python
#向量的点积或内积
v = np.array([2,1])
s = np.array([-3,2])

#计算点积
d = v @ s
#对应位置乘积的和 2 * (-3) + 1 * 2
print("点积是:",d)
```

```python
#计算两个向量的夹角
import math

v = np.array([2,1])
s = np.array([-3,2])

#计算向量的长度
vMag = np.linalg.norm(v)
sMag = np.linalg.norm(s)

print("vMag:",vMag)
print("sMag:",sMag)

#计算cosine(θ)
cos = (v @ s) /(vMag * sMag)
# 计算θ
theta = math.degrees(math.acos(cos))

print(theta)
```
