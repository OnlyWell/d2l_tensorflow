```python
'''
    微分和积分
'''
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False


#样本点
x = np.array(range(0, 11,1))
y = 2 * x + 5

#绘图
plt.grid()
plt.plot(x, y, color='g', linewidth=2, linestyle='--')
plt.xlim(-1,10)
plt.ylim(0,30)

#截距
x = [0,0]
y = [0,5]
plt.plot(x, y, color='r', linewidth=2, linestyle='--')
plt.annotate('截距', xy=(0,3), xytext=(1, 2), xycoords='data',arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=18)

#斜率
x = np.array([5,6])
y = 2 * x + 5
plt.plot(x, y, color='r', linewidth=2, linestyle='--')
plt.annotate('斜率', xy=(6,15), xytext=(8, 15), xycoords='data',arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=18)
plt.show()
```
