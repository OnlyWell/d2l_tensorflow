```python
'''
    偏微分,梯度下降法(Gradient Descent)
'''
# f = x^2 + y^2
from sympy import *

x, y = symbols('x y')

f = x ** 2 + y ** 2

df_dx = f.diff(x)
df_dy = f.diff(y)

print(df_dx)
print(df_dy)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#函数f(x) = x^2
def f(x):
    return x ** 2

#一阶导数
def df(x):
    return 2 * x

def GD(x_start,df,epochs, lr):
    #创建一个长度为epochs + 1的数组，用于存储每一轮迭代得到的权重值
    xs = np.zeros(epochs + 1)
    w = x_start
    xs[0] = w
    for i in range(epochs):
        dx = df(w)
        #权重的更新
        #W_new = W_old - learning_rate * dx
        w = w - lr * dx
        #将最新的权重存入到数组
        xs[i + 1] = w
    return xs

#超参数
x_start = 5 #起始权重
epochs = 25 #执行周期数
lr = 0.1 #学习率

# 执行梯度下降,df直接当参数传入
w = GD(x_start, df, epochs, lr)
#显示每一执行周期得到的参数
print(np.around(w,2))

#画图
color = 'r'
from numpy import arange

t= arange(-6.0,6.0,0.01)
plt.plot(t,f(t),c = 'b')
plt.plot(w,f(w),c = color, label = 'lr={}'.format(lr),marker = 'o')
plt.scatter(w,f(w),c = color)

#设定中文字型
from matplotlib.font_manager import FontProperties

font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14)
plt.title(u"梯度下降法",fontproperties = font)
plt.xlabel(u"x",fontproperties = font)
plt.ylabel(u"f(x)",fontproperties = font)

#矫正负号
plt.rcParams['axes.unicode_minus'] = False
plt.show()
```
